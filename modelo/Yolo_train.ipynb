```python
model = YOLO("yolov8x-seg.pt")
```

## Cargar el modelo preentrenado:
Los diferentes modelos preentrenados estan disponibles en [YOLOv8 GitHub](https://github.com/ultralytics/ultralytics)

```python
model.train(data = "DH_train.yaml", 
            epochs=200,
            cache = True, 
            imgsz=1024, # tamaÃ±o de las imgs
            batch=-1, # -1 == autobatch
            
            # Image augmentation
            hsv_h = 0.015,
            hsv_s=0.7,
            hsv_v = 0.4,
            scale = 0.5,
            flipud = 0.5,
            fliplr = 0.5,
            mixup = 0.5
            )
```
